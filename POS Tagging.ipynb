```python
import spacy
sp = spacy.load('en_core_web_sm')
```

```python
sen = sp(gone_w_the_wind)
i=0
for word in sen:
    print(f'{word.text:{i}} {word.pos_:{i}}')
    i+=1
```

While the initial pos_tagger using nltk looked to be accurate, the results from spacy library were more specific in capturing proper nouns.

```python
sen = sp(rumi)
i=0
for word in sen:
    print(f'{word.text:{i}} {word.pos_:{i}}')
    i+=1
```

The miscategorization of light in the Rumi quote happened again using the spacy library. Although it would be difficult to say how light is being used, as it could be categorized as either a verb, adjective or adverb in this context, depending on how you interpreted the quote.
