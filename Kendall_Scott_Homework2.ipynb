```python
years_5_8_McGuffey=years_5_8_McGuffey[682:years_5_8_McGuffey.index('End of the Project Gutenberg')]

years_9_12_NatureMyths=years_9_12_NatureMyths[682:years_9_12_NatureMyths.index('END OF THIS PROJECT GUTENBERG')]

years_6_11_Animal=years_6_11_Animal[682:years_6_11_Animal.index('END OF THIS PROJECT GUTENBERG')]
```

```python
#creating vocabulary lengths
years_5_8_McGuffey_vocab = word_tokenize(years_5_8_McGuffey)
print (years_5_8_McGuffey_vocab)
```

```python
years_6_11_Animal_vocab = word_tokenize(years_6_11_Animal)
print (years_6_11_Animal_vocab)
len(years_6_11_Animal_vocab)
```

```python
years_5_8_McGuffey_vocab = word_tokenize(years_5_8_McGuffey)
print (years_5_8_McGuffey_vocab)
len(years_5_8_McGuffey_vocab)
```

```python
years_5_8_McGuffey_vocab[0]
```

```python
i=0
McGlongwords=[]
while i<len(years_5_8_McGuffey_vocab):
  if len(years_5_8_McGuffey_vocab[i])>6:
    McGlongwords.append(years_5_8_McGuffey_vocab[i])
  i+=1
```

```python
len(McGlongwords)/len(years_5_8_McGuffey_vocab)
```

```python
i=0
Natlongwords=[]
while i<len(years_9_12_NatureMyths_vocab):
  if len(years_9_12_NatureMyths_vocab[i])>6:
    Natlongwords.append(years_9_12_NatureMyths_vocab[i])
  i+=1
```

```python
i=0
Animallongwords=[]
while i<len(years_6_11_Animal_vocab):
  if len(years_6_11_Animal_vocab[i])>6:
    Animallongwords.append(years_6_11_Animal_vocab[i])
  i+=1
```

```python
d = {'Book': ['years_5_8_McGuffey', 
              'years_6_11_Animal', 
              'years_9_12_NatureMyths'], 
     'large_words': [len(McGlongwords)/len(years_5_8_McGuffey_vocab), 
                     len(Natlongwords)/len(years_9_12_NatureMyths_vocab),
                     len(Animallongwords)/len(years_6_11_Animal_vocab)
                     ]}
vocab_df = pd.DataFrame(data=d)
```

```python

result = pd.concat([vocab_df, reading_df], axis=1)
```

```python
result['avg'] = result.mean(axis=1)
```

```python
result
```

```python
result_num=result["avg"]

result_num = result.loc[:, result.columns != 'Book']
```

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()
result_Scaled= Scaler.fit_transform(result_num)
result_Scaled = pd.DataFrame(result_Scaled, columns = result_num.columns)
result_Scaled
```

```python
d = {'Book': ['years_5_8_McGuffey', 
              'years_6_11_Animal', 
              'years_9_12_NatureMyths'], 
     'Vocabulary_Score': result_Scaled["avg"]}
vocab_df = pd.DataFrame(data=d)
```
